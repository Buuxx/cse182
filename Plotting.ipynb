```python
# Histogram: sliver length distribution

import matplotlib.pyplot as plt

input = open("sliverseqs.txt", "r")
length = []
for line in input:
    length.append(len(line.strip()))

fig, ax = plt.subplots()
ax.hist(length, 15, color="firebrick", edgecolor = "white")
plt.title("Sliver Length distribution")
plt.ylabel("Number of slivers")
plt.xlabel("Length (bp)")
plt.show()
```

```python
# Histogram: GC Content Distribution Among Slivers

import matplotlib.pyplot as plt
input = open("sliverseqs.txt", "r")
gc = ["G", "g", "C", "c"]
gccontent = []
for line in input:
    gccount = 0
    length = len(line.strip())
    for char in line.strip():
        if char in gc:
            gccount += 1
    gccontent.append(gccount/length)

# NEED TO ADD:
# Number of slivers hits a GC content larger then threshold?
# basic stats: mean, sd, ... (?)

fig, ax = plt.subplots()
ax.hist(gccontent, 15, color="mediumblue", edgecolor = "white")
plt.title("GC Content Distribution Among Slivers")
plt.xlabel("GC Content")
plt.ylabel("Number of Slivers")
plt.show()
```
